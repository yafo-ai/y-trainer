```python
from abc import ABC, abstractmethod

# 定义一个抽象基类
class MyBaseClass(ABC):

    @abstractmethod
    def must_implement_me(self):
        """子类必须实现这个方法"""
        ...

# 正确的子类：实现了抽象方法
class GoodChild(MyBaseClass):
    def must_implement_me(self):
        print("我实现了必须的方法！")

# 错误的子类：没有实现抽象方法
class BadChild(MyBaseClass):
    pass

# 测试
good = GoodChild()  # 正常，因为实现了 must_implement_me
good.must_implement_me()  # 输出：我实现了必须的方法！

# 下面这行会抛出异常，因为 BadChild 没有实现 must_implement_me
# bad = BadChild()  # TypeError: Can't instantiate abstract class BadChild with abstract method must_implement_me
```

```python
from openpyxl import load_workbook
import json

# 加载 Excel 文件
file_path = '/root/autodl-fs/auto-fs/y-trainer/dataset/mulit_hop.xlsx'
wb = load_workbook(filename=file_path)

# 获取所有工作表(sheet)名称
sheet_names = wb.sheetnames
print("工作表名称：", sheet_names)

# 选择第一个工作表（或者通过名称选择，如 wb['Sheet1']）
sheet = wb[sheet_names[0]]  # 或者 wb.active 获取当前活动的工作表

counter = 0

unified_data = []

# 遍历行数据
for row in sheet.iter_rows(values_only=True):  # values_only=True 表示返回单元格的值而不是单元格对象
    print(row)
    
    if counter != 0:
        
        unified_data.append({
            "ID": row[0],
            "input": row[1],
            "output": ''
        })
    
    counter += 1

with open(f'/root/autodl-fs/auto-fs/y-trainer/dataset/RL_task_251022.json', 'w', encoding='utf-8') as f:
    json.dump(unified_data, f, ensure_ascii=False, indent=2)


```
